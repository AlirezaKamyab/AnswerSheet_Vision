```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('./paper4_checked.png')
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 33, 33)
thresh = cv2.morphologyEx(thresh, cv2.MORPH_CROSS, (5, 5))
plt.imshow(thresh, cmap='gray')
```

```python
contours, hierarchies = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# NEXT, PREVIOUS, FIRST_CHILDREN, PARENT
```

```python
copyimg = img.copy()
i = 192
copyimg = cv2.drawContours(copyimg, contours, i, (255, 0, 0), 2)
plt.imshow(copyimg)
```

```python
x1, y1, w, h = cv2.boundingRect(contours[i])
x2, y2 = x1 + w, y1 + h
cropped = img[y1:y2, x1:x2, :]
```

```python
plt.imshow(cropped)
```
